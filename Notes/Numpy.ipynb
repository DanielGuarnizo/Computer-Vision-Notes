## <font color = yellow> Array concatenation 

```python
x= np.array(range(1,4))
y= np.array(range(4,7))
# function given a sequence of arrays it concatenate 
z = np.concatenate([x,y])
print(z)
```

```python
#x = np.array([range(i,i+3) for i in [1,4]])
x = np.array([[1,2,3],[4,5,6]])
#[[1 2 3]
# [4 5 6]]
y = np.array([[9,9,9],[91,91,91]])
# vstack --> 
# hstack -->
# dstack --> 
z = np.hstack([x,y])
print(z)
```

```python
np.vstack([x,y])
```

```python
x = np.array([1,2,3,4,5,6])
# np.split(array, *)  *indexes in which the array will be split , in our x1 will go until 2 index not compreset 
x1, x2, x3 = np.split(x,[2,4])
print(x1,x2,x3)
```

```python
x = np.arange(16).reshape((4,4))
upper, lower = np.split(x,[2])
print(f"Upper: \n{upper} \n Lower: \n{lower}")
```

```python
# thos method is a view of the data , that means if we modified it it will also affect the original one 
left, right = np.hsplit(x,[2])

# correct way to copy data 
print(f"Left: \n{left}\n Right: \n {right}")
```

```python
# correct way to copy data 
left = left.copy() 
left[0,0]= 333
print(x)
```

```python
right[0,0]= 333
print(x)
```
